```python
def noPageScrapper(bs_page, result_list):
    
    infos = bs_page.find_all('h4',class_='entry-title')
    prices = bs_page.find_all('span',class_='price')

    for i in range(len(prices)):

        item = {}

        name = infos[i].find('a')['title']
        link = infos[i].find('a')['href']
        price = prices[i].get_text()

        item['name'] = name
        item['link'] = link
        item['price'] = price
        item['store'] = 'New Order'

        result_list.append(item)


```

```python
def soupPage(link):
    page_html = requests.get(link).content  
    return BeautifulSoup(page_html,'lxml')
```
